Dalam olahraga air (Aquatics), kota mana yang memiliki rata-rata perolehan medali tertinggi?

```python
# memfilter data yang hanya berisi olahraga akuatik
summer = summer[summer['Sport'].str.contains("Aquatics")]

# melakukan agregasi data berdasarkan kolom city dan menghitung rata-rata perolehan medal
grouped = summer.groupby(['Sport','City','Medal']).size().reset_index(name='counts')

# menentukan kolom sport dengan city mana yang memiliki rata-rata perolehan medal tertinggi
grouped = grouped.sort_values(by='counts', ascending=False)

print("Kota dengan rata-rata perolehan medal tertinggi pada olahraga akuatik:")
print(grouped.head(1))
```

### **Soal 4 One Hot Encoding**

Buatlah **sebuah fungsi** bernama `one_hot`, yang menerima input sebuah numpy array.

Ex:<br>
input array [2,3,1,2] maka hasil dari fungsi kalian:

array [[0,1,0], [0,0,1], [1,0,0], [0,1,0]]

Constrain: <br>

- Integers dalam input selalu di awali 1. 
- Integers dalam input hanya positif dan bukan 0.
- Integers dalam input tidak akan skip angka.

```python
import numpy as np

def one_hot(arr):
    # mendapatkan jumlah unik dari input array
    uniq = np.unique(arr)
    n_classes = uniq.shape[0]
    
    # inisialisasi array one-hot encoding
    one_hot_encoded = np.zeros((arr.shape[0], n_classes), dtype=int)
    
    # mengisi nilai one-hot encoding sesuai dengan input array
    for i, value in enumerate(arr):
        index = np.where(uniq == value)[0][0]
        one_hot_encoded[i, index] = 1
        
    return one_hot_encoded
```

```python
input_array = np.array([3,2,1,2,3,3,3])
output_array = one_hot(input_array)

print(output_array)
```
