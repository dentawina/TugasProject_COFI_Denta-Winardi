```python
#PENJELASAN DATASET
Dataset Usairpollution ini terdiri dari 9 kolom
1. city : kota
2. So2 : kandungan sulfur dioksida mg/m^3
3. Temp : suhu tahunan rata rata (F)
4. manu : jumlah perusahaan manufaktur yang memperkerjakan 20 pekerja atau lebih
5. popul : ukuran populasi (sensus 1970) dalam ribuan
6. wind : kecepatan angin tahunan rata rata (mil/jam)
7. precip : curah hujan tahunan rata rata (inci)
8. predays : jumlah hari rata rata dengan curah hujan tahunan
9. klaster :
    
Dalam algoritma K-means, jarak antara data dan centroid dapat ditemukan Setelah jarak ditemukan,
setiap data akan ditempatkan pada cluster yang memiliki centroid terdekat.
Ketentuan K cluster (biasanya K adalah bilangan bulat positif) mempengaruhi jumlah centroid yang ada dan jumlah cluster.
Masing-masing cluster berdasarkan jarak antara data dan centroid. Jika data memiliki jarak terdekat dengan centroid C1,
maka data tersebut akan ditempatkan pada cluster 1. Begitu juga dengan centroid C2, C3, dan seterusnya.
Dengan demikian, setiap cluster adalah sekumpulan data yang memiliki jarak terdekat dengan centroid pada cluster tersebut.
Algoritma K-means akan mengulangi langkah-langkah seperti dijelaskan sebelumnya hingga hasil tidak berubah lagi,
sehingga setiap cluster akan berisi data yang serupa berdasarkan jarak antara data dan centroid.



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('USairpollution_quiz2.csv')
df
```

```python
klaster =df['clusters'].value_counts()
klaster
#klaster 0 ada 5 data
#klaster 1 ada 17 data
# klaster 2 ada 2 data
#klaster 3 ada 10 data
#klaster 4 ada 7 data
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['City'], df['clusters'])
ax.set_xlabel('City')
ax.set_ylabel('clusters')
plt.xticks(rotation=90)
plt.show()

#hasil berbeda dari tiap kluster untuk berbagai kota
```

```python
df.corr()['clusters'].sort_values(ascending=False)
#mencari korelasi dari setiap variabel terhadap "cluster"
```

```python
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap='BrBG',annot=True)

#korelasi tinggi pada kolom wind,precip,manu dan SO2
#karena korelasi tinggi pada 4 variabel tersebut kita mencoba untuk visualisasi untuk mencari kesamaan atau hubungan pola terhadap variabel kluster
```

```python
#To create plot using manu and SO2 by City
sns.FacetGrid(df, hue="City", size=5) \
   .map(plt.scatter, "SO2", "manu") \
   .add_legend()
plt.show()
#melihat persebaran SO2 DAN manu BERDASARKAN KOTA untuk melihat persebaran daerah mana yang memiliki kesamaan
```

```python
#To create plot using manu and SO2 by Cluster
sns.FacetGrid(df, hue="clusters",size=5) \
   .map(plt.scatter, "SO2", "manu") \
   .add_legend()
plt.show()
#melihat persebaran dari variabel SO2 dAN manu dilihat dari clusternya terlihat memiliki bebereapa kecocokan untuk 2 variabel ini
```

```python
#To create a plot using wind and precip by city
sns.FacetGrid(df, hue="City", size=5) \
   .map(plt.scatter, "wind",'precip') \
   .add_legend()
plt.show()
#melihat persebaran wind DAN precip BERDASARKAN KOTA untuk melihat persebaran daerah mana yang memiliki kesamaan
```

```python
#To create a plot using wind and precip by cluster
sns.FacetGrid(df, hue="clusters", size=5) \
   .map(plt.scatter, "wind",'precip') \
   .add_legend()
plt.show()
#melihat persebaran dari variabel wind dan precip dilihat dari clusternya terlihat memiliki bebereapa kecocokan untuk 2 variabel ini
```
